## 1. Operaciones Element-Wise (Elemento a Elemento)

Si tienes dos matrices del mismo tama침o, las operaciones (`+`, `-`, `*`, `/`) se aplican a cada pareja de n칰meros en la misma posici칩n.

```python
import numpy as np

# Creamos dos matrices 2x2
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[10, 20],
              [30, 40]])

print(f"Matriz A:\n{A}")
print(f"Matriz B:\n{B}")

# 1. SUMA DIRECTA
# No hace falta bucle. Suma el 1 con el 10, el 2 con el 20...
suma = A + B
print(f"\nSuma (A + B):\n{suma}")

# 2. MULTIPLICACI칍N POR ESCALAR
# Multiplicamos toda la matriz A por 10
escalado = A * 10 # TODO: Multiplica por 10
print(f"\nEscalado (A * 10):\n{escalado}")

# 3. CUADRADO
# Elevamos cada n칰mero al cuadrado
cuadrado = A ** 2 # TODO: Eleva a la potencia de 2
print(f"\nCuadrado (A^2):\n{cuadrado}")
```

## 2. Broadcasting (La Magia de NumPy)

쯈u칠 pasa si intentas sumar una Matriz (2D) y un Vector (1D)?
En matem치ticas cl치sicas: **Error**.
En NumPy: **Broadcasting**.

NumPy "estira" autom치ticamente el vector peque침o para que coincida con el grande.

```python
# Tenemos una matriz de (3 filas x 3 columnas)
matriz = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

# Tenemos un vector fila (1 fila, 3 columnas)
fila = np.array([0, 10, 20])

print(f"Matriz shape: {matriz.shape}")
print(f"Fila shape:   {fila.shape}")

# INTENTO DE SUMA
# NumPy va a coger la 'fila' y la va a repetir en cada fila de la matriz
resultado = matriz + fila # TODO: Suma la variable 'fila'
print(f"\nResultado Broadcasting:\n{resultado}")

# F칤jate: Ha sumado 0 a la primera columna, 10 a la segunda y 20 a la tercera.
# 춰En todas las filas a la vez!
```

## 3. M치scaras Booleanas (Filtros L칩gicos)

Esta es la herramienta m치s usada en Ciencia de Datos para limpiar valores.
En lugar de decir *"Dame la fila 3"*, decimos *"Dame los datos que sean mayores que 0"*.

1.  Creamos una **Pregunta** (`data > 0`) -> Devuelve `True`/`False`.
2.  Usamos esa respuesta como **칈ndice** -> Devuelve los datos.

```python
# Datos de temperatura (algunos son errores, -999)
datos = np.array([25, 30, -999, 28, -999, 32])

# PASO 1: LA M츼SCARA (La Pregunta)
# 쯈u칠 datos son v치lidos (mayores que -100)?
# Esto crea un array de Booleanos (True/False)
mascara = datos > -100 # TODO: Pon la condici칩n (-100)

print(f"Datos originales: {datos}")
print(f"M치scara l칩gica:   {mascara}")

# PASO 2: EL FILTRADO (Aplicar la m치scara)
# NumPy solo deja pasar los que son True
datos_limpios = datos[mascara] # TODO: Pon la variable 'mascara' dentro de los corchetes

print(f"Datos limpios:    {datos_limpios}")

# BONUS: 쮺u치ntos datos v치lidos tenemos?
# En Python, True cuenta como 1 y False como 0.
conteo = np.sum(mascara)
print(f"N칰mero de lecturas v치lidas: {conteo}")
```
