## 1. Manipulaci칩n "Hardcore" de Texto

Imagina que recibes este log de error de un servidor. Est치 sucio, tiene espacios y s칤mbolos que molestan. Vamos a limpiarlo paso a paso.

```python
# Dato sucio real
log_crudo = "   [CRITICAL] Error: 404 | User: Admin   \n"

# 1. LIMPIEZA DE EXTREMOS (.strip)
# Elimina los espacios en blanco y saltos de l칤nea del principio y final.
log_limpio = log_crudo.strip() # TODO: Usa el m칠todo de limpieza

print(f"Original: '{log_crudo}'")
print(f"Limpio:   '{log_limpio}'")

# 2. REEMPLAZO (.replace)
# Queremos quitar los corchetes para dejar el texto limpio.
# Truco: Reemplazamos "[" por nada ""
sin_corchetes = log_limpio.replace("[", "").replace("]", "") # TODO: Encadena otro replace

print(f"Sin corchetes: '{sin_corchetes}'")
```

## 2. Dividir y Trocear (Split & Slicing)

Ahora que est치 limpio, queremos separar la informaci칩n.
*   **Split:** Divide una frase en una **lista** de palabras usando un separador.
*   **Slicing:** Usa 칤ndices `[inicio : fin]` para recortar.

```python
# Texto actual: "CRITICAL Error: 404 | User: Admin"

# 1. DIVIDIR (.split)
# Queremos separar el Error del Usuario. El separador es la barra vertical "|".
partes = sin_corchetes.split("|") # TODO: Usa split con el separador correcto

print(f"Lista generada: {partes}")
print(f"Parte 1: {partes[0]}")

# 2. SLICING (El bistur칤)
# Imagina un c칩digo estandarizado: "2024-ID-999"
codigo = "2024-ID-999"

# Queremos solo el a침o (los primeros 4 caracteres)
anio = codigo[0:3] # TODO: Define el 칤ndice final

# Queremos los 칰ltimos 3 caracteres (el ID)
# Pista: Usa 칤ndices negativos para contar desde el final
id_num = codigo[-3:] # TODO: Desde el -3 hasta el final

print(f"A침o: {anio}, ID: {id_num}")

# TRUCO PRO: Invertir texto
# Sintaxis: [inicio : fin : paso]
print(f"Invertido: {codigo[::-1]}")
```

## 3. Funciones Lambda (Funciones An칩nimas)

En Data Science, a menudo necesitamos una funci칩n peque침a para usarla una sola vez (por ejemplo, para limpiar una columna de datos).

*   **Funci칩n Normal (`def`):** Varias l칤neas, tiene nombre.
*   **Lambda:** Una l칤nea, sin nombre. `lambda entrada: salida`

```python
# --- COMPARATIVA ---

# Opci칩n A: Funci칩n Cl치sica
def duplicar(x):
    return x * 2

# Opci칩n B: Lambda
# Sintaxis: lambda variable: operaci칩n
duplicar_lambda = lambda x: x * 2 # TODO: Completa la operaci칩n

# Probamos
print(f"Cl치sica: {duplicar(5)}")
print(f"Lambda:  {duplicar_lambda(5)}")

# --- CASO DE USO REAL: PROCESAR UNA LISTA ---
precios_sucios = ["100$", "200$", "50$"]

# Queremos quitar el "$" y convertir a entero.
# Usamos `map` para aplicar una lambda a toda la lista.

# La lambda recibe un precio 'p', le quita '$' y lo convierte a int
precios_limpios = list(map(lambda p: int(p.replace("$", "")), precios_sucios)) # TODO: Usa replace dentro de la lambda

print(f"Precios num칠ricos: {precios_limpios}")
# 춰Hemos limpiado una lista entera en una l칤nea de c칩digo!
```
