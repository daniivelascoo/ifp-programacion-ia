## 1. El Histograma (La Forma)

Vamos a usar el dataset de propinas (`tips`).
Queremos saber: **Â¿CuÃ¡nto suele pagar la gente en total?**
Â¿Es comÃºn pagar 10\$? Â¿Es comÃºn pagar 50\$?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos datos de ejemplo (Propinas en un restaurante)
df = sns.load_dataset("tips")

print("--- Primeras filas ---")
display(df.head())

# CREAR HISTOGRAMA
plt.figure(figsize=(8, 5))

# kde=True aÃ±ade la lÃ­nea curva suave (Kernel Density Estimate)
# bins=20 define cuÃ¡ntas barras queremos
sns.histplot(data=df, x='total_bill', kde=True, bins=20, color='teal') # TODO: Usa la funciÃ³n histplot

plt.title("DistribuciÃ³n de la Cuenta Total")
plt.xlabel("Total Cuenta ($)")
plt.show()

# REFLEXIÃ“N:
# FÃ­jate en la "cola" hacia la derecha.
# La mayorÃ­a paga entre 10 y 20, pero hay gente que paga 50.
# Esto NO es una distribuciÃ³n Normal perfecta.
```

## 2. El Boxplot (El Cazador de AnomalÃ­as)

El Histograma estÃ¡ bien, pero el **Boxplot** (Diagrama de Caja y Bigotes) es tÃ©cnico y preciso.
Nos dice:
1.  **Mediana:** La lÃ­nea del centro de la caja.
2.  **El 50% central:** La caja (Rango IntercuartÃ­lico).
3.  **Los Outliers:** Los puntos que salen fuera de los "bigotes".

```python
plt.figure(figsize=(8, 5))

# Queremos ver la distribuciÃ³n de propinas ('tip')
# OrientaciÃ³n: Si pones x='tip', sale horizontal. Si pones y='tip', sale vertical.
sns.boxplot(data=df, x='tip', color='orange') # TODO: Usa la funciÃ³n boxplot

plt.title("AnÃ¡lisis de Propinas (Detectando tacaÃ±os y generosos)")
plt.show()

# FÃJATE EN LOS PUNTOS NEGROS A LA DERECHA:
# Esos son los Outliers. Propinas de 6$, 8$, 10$...
# Son casos raros que podrÃ­an confundir a una IA.
```

## 3. Comparando CategorÃ­as

El poder real del Boxplot es comparar grupos.
*   *Â¿QuiÃ©n gasta mÃ¡s, los Fumadores o los No Fumadores?*
*   *Â¿QuÃ© dÃ­a de la semana es mÃ¡s caro?*

```python
plt.figure(figsize=(10, 6))

# Eje X: CategorÃ­a (DÃ­a de la semana)
# Eje Y: NumÃ©rico (Total Cuenta)
# Palette: Colores bonitos ('Set2', 'viridis', 'magma')

sns.boxplot(data=df, x='day', y='total_bill', hue='day', palette='Set3', legend=False) # TODO: Columna 'total_bill'

plt.title("Gasto por DÃ­a de la Semana")
plt.show()

# PREGUNTA VISUAL:
# Mira los grÃ¡ficos. Â¿QuÃ© dÃ­a tiene los "bigotes" mÃ¡s largos (mÃ¡s variabilidad)?
# Â¿QuÃ© dÃ­a tiene la mediana (lÃ­nea interna) mÃ¡s alta?
```
