## 1. Series y DataFrames

*   **Series:** Es una sola columna de datos. (1D)
*   **DataFrame:** Es una colecci贸n de Series. Una tabla. (2D)

Vamos a crear uno manualmente para entenderlo.

```python
import pandas as pd
import numpy as np

# 1. CREAR UN DICCIONARIO (Datos crudos)
datos = {
    'cancion': ['Despacito', 'Bohemian Rhapsody', 'Imagine', 'Baby Shark'],
    'artista': ['Luis Fonsi', 'Queen', 'John Lennon', 'Pinkfong'],
    'reproducciones_mill': [8000, 2000, 500, 10000],
    'es_bailable': [True, True, False, True]
}

# 2. CONVERTIR A DATAFRAME
# La funci贸n pd.DataFrame() convierte el diccionario en una tabla bonita
df = pd.DataFrame(datos) # TODO: Completa el constructor DataFrame

# 3. VISUALIZAR
# En Colab, no hace falta print(). Si pones la variable al final, se ve mejor.
display(df)

print(f"Tipo de objeto: {type(df)}")
```

## 4. El Dilema: LOC vs ILOC

Esta es la parte que m谩s confunde al principio. 驴C贸mo selecciono datos?

*   **`iloc` (Integer Location):** Busca por **Posici贸n** (como NumPy). "Dame la fila 0".
*   **`loc` (Location):** Busca por **Etiqueta** o **Condici贸n**. "Dame los productos con precio > 100".

```python
# Usamos df_tienda

# 1. USANDO ILOC (Posici贸n)
# Queremos la primera fila (铆ndice 0)
producto_1 = df_tienda.iloc[0] # TODO: Pon un 0
print(f"Producto 1 (por posici贸n): {producto_1['producto']}")

# 2. USANDO LOC (Condici贸n L贸gica - La potencia real)
# Queremos filtrar: Solo los productos de 'Electr贸nica'
# Sintaxis: df.loc[ CONDICIN ]

filtro = df_tienda['categoria'] == 'Electr贸nica'
electronica = df_tienda.loc[filtro] # TODO: Pon la variable filtro dentro

print("\nSolo Electr贸nica:")
display(electronica)

# 3. EXTRAER COLUMNA
# Queremos solo la columna de precios (esto devuelve una Series)
precios = df_tienda['precio'] # TODO: Pon el nombre de la columna precio
print(f"\nLista de precios:\n{precios.values}")
```
