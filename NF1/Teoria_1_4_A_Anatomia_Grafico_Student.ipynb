## 1. Preparaci√≥n de Datos

Vamos a crear un peque√±o dataset de ventas mensuales para dibujarlo.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Datos: Ventas vs Costes en el primer cuatrimestre
df = pd.DataFrame({
    'mes': ['Ene', 'Feb', 'Mar', 'Abr'],
    'ventas': [100, 150, 130, 200],
    'costes': [80, 90, 85, 100]
})

display(df)
```

## 2. Matplotlib (La Forma Manual)

Matplotlib funciona con una jerarqu√≠a:
1.  **Figure (Fig):** Es el lienzo en blanco (la ventana).
2.  **Axes (Ax):** Es el gr√°fico en s√≠ (los ejes X e Y donde pintamos).

Usamos `plt.subplots()` para crear ambos a la vez.

```python
# 1. CREAR LIENZO Y EJES
# figsize=(ancho, alto) en pulgadas
fig, ax = plt.subplots(figsize=(6, 4))

# 2. DIBUJAR DATOS
# Pintamos las Ventas (L√≠nea Azul)
# x = columna mes, y = columna ventas
ax.plot(df['mes'], df['ventas'], label='Ingresos', color='blue', marker='o')

# Pintamos los Costes (L√≠nea Roja Discontinua)
# usa linestyle='--'
ax.plot(df['mes'], df['costes'], label='Gastos', color='red', linestyle='--') # TODO: Rojo y Discontinuo

# 3. PERSONALIZAR (Etiquetas)
ax.set_title("Evoluci√≥n Financiera (Matplotlib)")
ax.set_xlabel("Meses")
ax.set_ylabel("Euros (‚Ç¨)")

# 4. MOSTRAR LEYENDA
# Muestra los 'label' que definimos arriba
ax.legend()

# 5. MOSTRAR RESULTADO
plt.show()
```

## 3. Seaborn (La Forma Moderna)

Seaborn est√° construido *encima* de Matplotlib.
La gran ventaja: **Entiende los DataFrames de Pandas**.
No tienes que pasarle `x` e `y` por separado; le pasas el DataFrame entero y le dices los nombres de las columnas.

```python
# 1. ACTIVAR ESTILO
# Seaborn tiene temas que mejoran la est√©tica autom√°ticamente
sns.set_theme(style="darkgrid") # TODO: Prueba 'darkgrid', 'whitegrid' o 'ticks'

# 2. CREAR GR√ÅFICO (Lineplot)
# data = el dataframe completo
# x = nombre de la columna para el eje X
# y = nombre de la columna para el eje Y

plt.figure(figsize=(6, 4))

# Pintamos Ventas
sns.lineplot(data=df, x='mes', y='ventas', label='Ingresos', marker='o') # TODO: Columna ventas

# Pintamos Costes
sns.lineplot(data=df, x='mes', y='costes', label='Gastos', linestyle='--') # TODO: Columna costes

plt.title("Evoluci√≥n Financiera (Seaborn)")
plt.show()

# Nota: ¬°F√≠jate que Seaborn pone los nombres de los ejes autom√°ticamente!
```

## 4. Cambiando el tipo de gr√°fico

Con Seaborn, cambiar de L√≠nea a Barras es cambiar una sola palabra: de `lineplot` a `barplot`.

```python
plt.figure(figsize=(6, 4))

# Queremos ver las Ventas en barras
sns.barplot(data=df, x='mes', y='ventas', hue='mes', palette='viridis', legend=False) # TODO: Usa la funci√≥n barplot

plt.title("Ventas por Mes")
plt.show()
```
