## 1. La Lista vs El Array

Una lista de Python puede guardar cualquier cosa `[1, "Hola", True]`. Eso la hace pesada.
Un **Array de NumPy** es un bloque de memoria contigua (como en C) donde todos los datos son del mismo tipo.

```python
# Importamos la librer√≠a est√°ndar con su alias universal
import numpy as np

# 1. LISTA DE PYTHON (Cl√°sica)
lista_py = [1, 2, 3]

# 2. ARRAY DE NUMPY (Nuevo)
# Convertimos la lista en un array
array_np = np.array(lista_py) # TODO: Completa el constructor (array)

print(f"Lista: {lista_py}")
print(f"Array: {array_np}")

# PISTA VISUAL:
# F√≠jate que el Array se imprime SIN COMAS entre los n√∫meros.
# Eso indica que es un vector matem√°tico, no una lista de objetos.
```

## 2. La Carrera de Velocidad (Vectorizaci√≥n)

Vamos a multiplicar por 2 una lista de **1 mill√≥n de n√∫meros**.
*   **Python:** Tiene que ir n√∫mero a n√∫mero, comprobar qu√© tipo es y multiplicar.
*   **NumPy:** Sabe que todo son n√∫meros y lo hace de golpe (SIMD).

```python
import time

# Preparamos el terreno: 1 Mill√≥n de datos
tama√±o = 1_000_000
datos_py = list(range(tama√±o))
datos_np = np.arange(tama√±o) # arange es el range de NumPy

print(f"üèéÔ∏è Iniciando carrera con {tama√±o} datos...")

# --- CORREDOR 1: PYTHON (Bucle For / List Comp) ---
inicio = time.time()
# Tenemos que recorrer la lista uno a uno
resultado_py = [x * 2 for x in datos_py]
fin = time.time()
print(f"Tiempo Python: {fin - inicio:.5f} segundos")

# --- CORREDOR 2: NUMPY (Vectorizaci√≥n) ---
inicio = time.time()
# Operamos sobre el array entero directamente
resultado_np = datos_np * 2 # TODO: Multiplica por 2 directamente
fin = time.time()
print(f"Tiempo NumPy:  {fin - inicio:.5f} segundos")

# CONCLUSI√ìN:
# NumPy suele ser entre 50 y 100 veces m√°s r√°pido.
```

## 3. Tipos de Datos (Dtypes) y Memoria

En Python, un n√∫mero entero (`int`) ocupa mucho espacio porque es un objeto complejo. En NumPy, podemos elegir cu√°ntos bits usar.
*   `int64`: Entero gigante (Est√°ndar).
*   `int8`: Entero peque√±o (-128 a 127). ¬°Ocupa 8 veces menos memoria!

```python
# Array est√°ndar (se adapta a los datos)
arr_float = np.array([1.5, 2.5, 3.5])
print(f"Tipo detectado: {arr_float.dtype}")

# FORZAR EL TIPO (Ahorro de memoria)
# Imagina que son edades (nadie tiene 300 a√±os, cabe en int8)
edades = np.array([10, 20, 90], dtype='int8') # TODO: Usa 'int8'

print(f"Tipo forzado: {edades.dtype}")

# ‚ö†Ô∏è PELIGRO DEL DESBORDAMIENTO (Overflow)
# Si usas int8, el m√°ximo es 127. ¬øQu√© pasa si metemos 200?
prueba_peligrosa = np.array([200], dtype='int8')
print(f"Valor 200 en int8 se convierte en: {prueba_peligrosa}")
# Explicaci√≥n: Da la vuelta al contador binario (Overflow). ¬°Cuidado en producci√≥n!
```
